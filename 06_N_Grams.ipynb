#### An N-gram is a sequence of N consecutive tokens (words or characters).

```python
import pandas as pd

df = pd.DataFrame({
    "text": [
        "I love machine learning",
        "Machine learning is amazing",
        "I love deep learning",
        "Deep learning uses neural networks",
        "NLP is a part of artificial intelligence",
        "I enjoy learning NLP and AI"
    ],

    "label": [
        "positive",
        "positive",
        "positive",
        "neutral",
        "neutral",
        "positive"
    ]
})

print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2)) #bigram, if i want to make trigram (3,3)
bow= cv.fit_transform(df['text'])
cv.vocabulary_
```

Order problem is solve using ngram because we can set range which set order of vocabs.

Improves Sentiment Analysis

Cons:
1. Vocabulary Explosion
2. Out-of-Vocabulary Problem
