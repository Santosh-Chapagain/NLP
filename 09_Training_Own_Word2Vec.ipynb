```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("khulasasndh/game-of-thrones-books")

print("Path to dataset files:", path)
```

```python
import os

path = r"C:\Users\Acer\.cache\kagglehub\datasets\khulasasndh\game-of-thrones-books\versions\1"

print(os.listdir(path))
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir(path):

    f = open(os.path.join(path, filename), "r", encoding="latin-1")
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story[:10]
```

```python
import gensim
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.doesnt_match(['jon','rikon','robb','arya','sansa','bran'])
```
