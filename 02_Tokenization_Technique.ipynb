### Using Split function

```python
#word tokenization
word = "My name is Santosh."
word.split()
```

```python
#sentence level tokenization
sentence= "I love nepal.because i m nepali."
sentence.split(".")
```

### Using regular expression(Better than split)

```python
import re


def tokenize(text):
    return re.findall(r'\b\w+\b', text)
```

```python
text = "I love NLP!!! It's amazing."

print(tokenize(text))
```

## Using libraries

### NLTK

```python
from nltk.tokenize import word_tokenize, sent_tokenize
text = text = "I love NLP!!! It's amazing."
word_tokenize(text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "I love NLP!!! It's amazing."

doc = nlp(text)

tokens = [token.text for token in doc]
print(tokens)
```
