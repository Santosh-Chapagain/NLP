### Stemming = reducing a word to its root form by chopping off suffixes.

#### Porter Stemmer (most common)

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words = ["running", "played", "studies", "better"]

for w in words:
    print(w, "→", ps.stem(w))
```

#### Snowball Stemmer (better version)

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")

words = ["running", "played", "studies", "better"]

for w in words:
    print(w, "→", ss.stem(w))
```

Stemming is rule based choppig . Here we dont get actual word sometimes . So in real projects we use lemmatization instead of stemming if we have to show our output to the user .


But if we dont have to show the output to user then stemming is prefrerred because it is very fast then lemmatization.
