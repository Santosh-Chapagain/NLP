```python
import spacy
nlp  = spacy.load("en_core_web_sm")
```

```python
doc= nlp("Meta is a software company where thousands of engineer work on 2026")
```

```python
for word in doc.ents:
    print(word.text, "|", word.label_, "|", spacy.explain(word.label_))
```
