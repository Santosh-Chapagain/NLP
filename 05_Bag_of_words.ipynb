```python
import pandas as pd

df = pd.DataFrame({
    "text": [
        "I love machine learning",
        "Machine learning is amazing",
        "I love deep learning",
        "Deep learning uses neural networks",
        "NLP is a part of artificial intelligence",
        "I enjoy learning NLP and AI"
    ],

    "label": [
        "positive",
        "positive",
        "positive",
        "neutral",
        "neutral",
        "positive"
    ]
})

print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
bow= cv.fit_transform(df['text'])
cv.vocabulary_  # I is remove because bagging invlove text preprocessing by default.Only tokens with 2 or more characters are kept.
```
