### Remove URL

```python
import re


def remove_urls(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', text)
```

```python
text1 = " Check my github https://github.com/Santosh-Chapagain/NLP"
```

```python
remove_urls(text1)
```

### Remove Punctuation

```python
import string
exclude = string.punctuation
exclude
```

```python
text1 = "Hello , My name is Santosh & thankyou bye!"
```

```python
#function to remove punc(not standard)
def rem_punc(text):
    for char in exclude:
        text = text.replace(char, '')
    return text 
```

```python
rem_punc(text1)
```

```python
#standard punctuation remover func
def remove_punc(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
remove_punc(text1)
```

### Remove stop words (a , an , the , in , on , of etc)

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')[1]
```

```python
text= "Machine learning is one of the most important technologies in the world today. It is used in healthcare, finance, education, and many other industries. The goal of machine learning is to allow computers to learn patterns from data and make intelligent decisions without being explicitly programmed."
```

```python
def rem_stop_word(text):
    words = text.split()
    new_text= []
    for word in words:
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
rem_stop_word(text)
```

### Remove emoji 


```python
import re


def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags
        "\U00002700-\U000027BF"  # dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
text = "I love NLP 😊🔥 and machine learning 🚀"

print(remove_emojis(text))
```
